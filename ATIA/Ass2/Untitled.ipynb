```python
import cv2
import numpy as np
import skimage
import matplotlib.pyplot as plt
```

```python
imA = cv2.imread('../I74.JPG')
imA = cv2.cvtColor(imA,cv2.COLOR_BGR2GRAY)
imA = skimage.filters.gaussian(imA,15)
plt.imshow(imA,cmap='gray')
```

```python
samplingSpace = 100
intsA = imA[::samplingSpace,::samplingSpace]
intsB = imA[::samplingSpace,::samplingSpace]

skimage.metrics.normalized_mutual_information(intsA,intsB,bins=256)
```
